```python
# delete the rows with loan_status as "Current"
loan = loan[loan["loan_status"].apply(lambda x:False if x == "Current" else True)]
loan["loan_status"].unique()
```

```python
# now convert it into numerical info
loan['loan_status']=loan['loan_status'].apply(lambda x: 0 if x=='Fully Paid' else 1 )
loan["loan_status"].unique()
```

```python
# create bins for loan_amnt range
binnings = [0, 5000, 10000, 15000, 20000, 25000, 36000]
loan_amnt_bukcet = ['0-5000', '5000-10000', '10000-15000', '15000-20000', '20000-25000','25000+']
loan['loan_amnt_range'] = pd.cut(loan['loan_amnt'], binnings, labels=loan_amnt_bukcet)
loan['loan_amnt_range'].unique()
```
